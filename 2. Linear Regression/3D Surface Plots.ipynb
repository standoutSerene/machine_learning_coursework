```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
a = [1,2,3]
b = [4,5,6,7]
a,b = np.meshgrid(a,b)
```

```python
print(a)
print(b)
```

```python
# 'a' will be repeated as a row, # elements in b times
# 'b' will be repeated as a column, # elements in a times
```

```python
# and both of them will have the same shape
```

```python
# pair up the corresponding points, a(i,j) with b(i,j)
```

```python
# this will basically be a flat surface, or points of a rectangular lattice(kind of a mesh)
```

```python
# we need to define the elevation at each of the points of the mesh
```

```python
fig = plt.figure()
axes = fig.gca(projection = '3d')
axes.plot_surface(a,b,a**2+b**2,cmap = 'rainbow')
```

```python
m = a.shape[0]
n = a.shape[1]
print(m,n)

for i in range(m):
    for j in range(n):
        print("(%d,%d)"%(b[i,j],a[i,j]),end = " ")
    print()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.arange(-4,4,0.02)
b = a
```

```python
a,b = np.meshgrid(a,b)
print(a)
print(b)
```

```python
m = a.shape[0]
n = a.shape[1]
print(m,n)

for i in range(m):
    for j in range(n):
        print("(%d,%d)"%(b[i,j],a[i,j]),end = " ")
    print()
```

```python
fig = plt.figure()
axes = fig.gca(projection = '3d')
axes.plot_surface(a,b,a**2+b**2,cmap = 'rainbow')
plt.show()
```

```python
# we can also have a contour plot
```

```python
fig = plt.figure()
axes = fig.gca(projection = '3d')
axes.contour(a,b,a**2+b**2, cmap = 'rainbow')        # you can also use 'coolwarm'
plt.show()
```

```python
# we can project this on a 2D surface and get a 2D contour plot
```

```python
plt.contour(a,b,a**2+b**2)
plt.show()
```
