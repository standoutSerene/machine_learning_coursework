## the big data machine :)
## gradient descent
- start with a random theta
- repeat untill convergence 
    - update theta according to the rule

```python
def hypothesis(x, theta):
    return theta[0] + theta[1]*x
# x is a scalar and theta is a vector
```

```python
def error(X,Y,theta):
    # X is the X :), Y is the actual Y
    m = X.shape[0]
    err = 0
    for i in range(m):
        err += (hypothesis(X[i],theta)-Y[i])**2
    return err 
```

```python
def gradient(X,Y,theta):
    grad = np.zeros((2,))
    m = X.shape[0]
    for i in range(m):
        hx = hypothesis(X[i],theta)
        grad[0] += (hx-Y[i]) * 1
        grad[1] += (hx-Y[i]) * X[i]
    return grad
    
        
    
```

```python
def gradientDescent(X,Y,learning_rate = 0.001):
    
    theta = np.zeros((2,))
    itr = 0
    max_itr = 100
    
    error_list = []
    theta_list = []
    
    while(itr<=max_itr):
        grad = gradient(X,Y,theta)
        e = error(X,Y,theta)
        error_list.append(e)
        theta_list.append((theta[0],theta[1]))
        theta[0] = theta[0] - learning_rate * grad[0]
        theta[1] = theta[0] - learning_rate * grad[1]
        itr += 1
        
        
    return theta,error_list,theta_list
        
```

```python
final_theta,error_list,theta_list = gradientDescent(X,Y)
```

```python
plt.plot(error_list)
plt.show()
```

## Visualizing Gradient Descent

```python
# plotting the error surface and contours
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
# we want to plot theta0, theta1, and J(theta)
```

```python
T0 = np.arange(-2,3,0.01)
T1 = np.arange(-2,3,0.01)

T0,T1 = np.meshgrid(T0,T1)
```

```python
J = np.zeros(T0.shape)
m = T0.shape[0]
n = T0.shape[1]

for i in range(m):
    for j in range(n):
        J[i,j] = np.sum((Y - T1[i,j]*X - T0[i,j])**2)
```

```python
# plot the J function
```

```python
fig = plt.figure()
axes = fig.gca(projection = '3d')
axes.plot_surface(T0,T1,J, cmap = 'rainbow')
plt.show()
```

```python
# we want to trace(plot) the trajectory of the gradient descent
```

```python
# for that, we require the values of theta0 and theta1 at every iteration
```

```python
# let's first make the contours
```

```python
fig = plt.figure()
axes = fig.gca(projection = '3d')
axes.set_xlim([-2,2])
axes.set_ylim([-2,2])
axes.contour(T0,T1,J,cmap = 'rainbow')
plt.title("Contours in 3D")
plt.show()
```

```python
# let's also make 2D contours
```

```python
plt.contour(T0,T1,J)
plt.title("2D contour")
plt.show()
```
