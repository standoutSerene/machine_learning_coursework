## More on Plotting

```python
x = np.array([1,2,3,4,5])
y = np.array([10,20,50,80,90])
plt.plot(x,x**2)
plt.plot(x,x**3)
plt.xlabel("no of days")
plt.ylabel("prices")
plt.title("stock price comparison")
plt.show()
```

```python
# suppose we want to label each of the curves corresponding to their respective years
```

```python
# or say their respective stock names
```

```python
x = np.array([1,2,3,4,5])
y = np.array([10,20,50,80,90])
plt.plot(x,x**2, label = '2018')
plt.plot(x,x**3, label = '2017')
plt.xlabel("no of days")
plt.ylabel("prices")
plt.title("stock price comparison")
plt.legend()
plt.show()
```

```python
# need to supply the labels, and call the legend function
```

## Bar Chart

```python
numbers = np.random.randint(0,10,5)
indices = np.arange(5)

numbers2 = np.random.randint(0,10,5)
indices2 = np.arange(5)


print(numbers)
print(indices)
```

```python
# the first two arguments in the randint function specify the range, and the third one specifies the
# number of numbers
```

```python
plt.bar(indices,numbers, 0.5, color = 'orange', label = 'Gold')
plt.bar(indices2, numbers2, 0.5, color = 'green', label = 'silver')
plt.legend()
plt.show()
```

```python
# 0.5 specifies the bar-width
```

```python
# note that the two are overlapping and it is not that they are on the top of each other
```

```python
# to make it look better, we can shift the indices of the second by 0.5
```

```python
plt.bar(indices,numbers, 0.25, color = 'orange', label = 'Gold')
plt.bar(indices2 + 0.25, numbers2, 0.25, color = 'red', label = 'silver')
plt.legend()
plt.show()
```

## Histogram
- estimate of the probability distribution of a continuous variable
- differs from bar graph
- bar graph relates two variables, histogram only one

```python
plt.hist(vals,50)
plt.show()
```

```python
# 50 is the number of intervals
```

```python
# what does it depict ?
# how many values fall in each interval or equally, the probability
```

```python
# on the x-axis, we have the points drawn from the normal distribution
# with u = 5, and sigma = 2
```

```python
# standard normal distribution is a normal distribution with u zero and sigma one
```
